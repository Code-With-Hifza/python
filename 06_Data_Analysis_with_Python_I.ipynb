## Dealing with Missing Values

## Replace ? in _normalized-losses_ column by NAN

```python
df.head()
```

```python
df["normalized-losses"]
```

```python
df["normalized-losses"].replace("?", np.nan, inplace = True)
```

```python
df.head()
```

```python
df["normalized-losses"] = pd.to_numeric(df["normalized-losses"]) # Convert this column from object to numeric column
```

```python
df.dtypes
```

```python
mean = df["normalized-losses"].mean()
mean
```

```python
df["normalized-losses"].replace(np.nan, mean, inplace=True) # replace nan with mean value # inplace=True (This will replace nan in the original records and will not create a new copy or column)
```

```python
df["normalized-losses"]
```
